```python
import nltk 
from nltk.tokenize import word_tokenize 
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
data = "I like AI and Math"

tokenization = word_tokenize(data)

result = nltk.pos_tag(tokenization)

result
```
